```python
# making one that works in any orther
# work in progress
import re
user_input = input('put your ingridients, followed by a comma')
user_input_list = user_input.split(',')

pattern = r'(?i)\b(?:' + '|'.join(user_input_list) + r')\b'

matched_recipes = recipes.loc[recipes['RecipeInstructions'].str.contains(pattern, flags=re.IGNORECASE, regex = True)]

result_df = pd.DataFrame(matched_recipes, columns = recipes.columns )
styled_df = result_df.style.set_properties(subset=pd.IndexSlice[::2], background='#f5f5f5')

print(styled_df)
```
